```python
#
#    Besvarelse P7.4
#
#  Printer gjennomsnitt av kolonner i fil

from pathlib import Path


def get_lines_from_file(filename, delimiter):
    path = Path(__file__).with_name(filename)

    with path.open("r") as f:
        lines = [line.strip().split(delimiter) for line in f.readlines()]
    
    lines = [[float(line[0]), float(line[1])] for line in lines]
    return lines


def main():
    print("Calculate average of columns. <file.txt> must contain floating")
    print("point numbers in two columns")
    filename  = input("Filename: ")
    delimiter = input("Delimiter: ")

    lines = get_lines_from_file(filename, delimiter)

    entries = len(lines)
    print(f"Avg. of col 1: {sum([val[0] for val in lines]) / entries}")
    print(f"Avg. of col 2: {sum([val[1] for val in lines]) / entries}")

    
if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.6
#
#  Finn streng i filer og print linjer med denne strengen


#!/usr/bin/python3
from pathlib import Path
from sys import argv
from colorama import Fore


def find_print_line(filenames, search_word):
    for filename in filenames:
        path = Path(__file__).with_name(filename)

        with path.open("r") as file:
            lines = file.readlines()
        
        for n, line in enumerate(lines):
            if search_word in line:
                
                # Formatting, adding red colour to the matched word
                line = line.strip("\n")
                match = Fore.RED + search_word + Fore.RESET
                line = line.replace(search_word, match)

                print(filename, "line", n, "-->", line)
                

def main():
    usage = """Usage: python find.py searchword file1 file2 file3"""
    if len(argv) == 1:
        quit()
    
    if argv[1] in ["h", "help"]:
        print(usage)
        quit()
        
    search_word = argv[1]
    files = argv[2:]
    
    find_print_line(files, search_word)
    
    
if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.9
#
#  Lager kopi av fil med reverserte linjer


#!/usr/bin/python3
from pathlib import Path
from sys import argv


def reverse_file(orig_file, new_file):
    old = Path(__file__).with_name(orig_file)
    new = Path(__file__).with_name(new_file)

    with old.open("r") as file:
        lines = file.readlines()

    lines.reverse()

    with new.open("w") as file:
        file.writelines(lines)


def main():
    usage = """Usage: python revlines.py original_file new_reverted_file"""
    
    if len(argv) != 2:
        print(usage)
        
    orig_file = argv[1]
    new_file  = argv[2]
    reverse_file(orig_file, new_file)
    
    print("Made a line reverted copy of", orig_file, "named", new_file)


if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.20
#
#  Krypterer eller dekrypterer fil


#!/usr/bin/python3
from pathlib import Path
import argparse


def get_text_from_file(filename):
    filepath = Path(filename)
    
    with filepath.open("r") as file:
        text = file.read()
    return text

        
def write_to_file(filename, text):
    filepath = Path(filename)
    
    with filepath.open("w") as file:
        file.write(text)


def apply_encrypt_or_decrypt(text, key, decrypt=False):
    """Encryption method: Random monoalphabet cipher
         for instance, with keyword 'feather', translation is as follows:
         
         feathrzyxwvusqponmlkjigdcb  FEATHRZYXWVUSQPONMLKJIGDCB
                     |                            |
                     v                            v
         zyxwvutsrqponmlkjihgfedcba  ZYXWVUTSRQPONMLKJIHGFEDCBA
         
       By default encrypts 'text'. Pass 'decrypt=True' to decrypt.
    """
    alphabet_list = [chr(x) for x in range(122, 96, -1)]
    alphabet_str  = "".join(alphabet_list)
    key = key.lower()

    cipher_from_key   = "".join(dict.fromkeys(key))
    cipher_from_alpha = "".join([c for c in alphabet_list if c not in set(key)])
    cipher = cipher_from_key + cipher_from_alpha
    
    ciph_upper_lower = cipher + cipher.upper()
    alpha_upper_lower = alphabet_str + alphabet_str.upper()

    if decrypt:
        translation_tab = text.maketrans(ciph_upper_lower, alpha_upper_lower)
    else:
        translation_tab = text.maketrans(alpha_upper_lower, ciph_upper_lower)
        
    processed_text = text.translate(translation_tab)
    return processed_text


def main():    
    ### Parse arguments
    parser = argparse.ArgumentParser(
        description="Encrypt or decrypt a file using a keyword.")
    
    action = parser.add_mutually_exclusive_group(required=True)
    action.add_argument("-d", "--decrypt",
                        action="store_true",
                        help="decrypt file")
    action.add_argument("-e", "--encrypt",
                        action="store_true",
                        help="encrypt file")
    
    parser.add_argument("-k", "--key",
                        required=True,
                        help="key for generating cipher (case insensitive)")
    parser.add_argument("-i", "--input",
                        required=True,
                        help="input file")
    parser.add_argument("-o", "--output",
                        required=True,
                        help="output file")
    
    args = parser.parse_args()
    inputfile  = args.input
    outputfile = args.output
    keyword    = args.key.upper()
   
    
    ### Read file
    text = get_text_from_file(inputfile)
    
    ### Encrypt or decrypt contents of file
    if args.decrypt:
        text = apply_encrypt_or_decrypt(text, keyword.upper(),
                                        decrypt=True)
    elif args.encrypt:
        text = apply_encrypt_or_decrypt(text, keyword.upper(),
                                        decrypt=False)
    
    ### Write to file
    write_to_file(outputfile, text)
    

if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.21
#
#  Printer prosentvis hyppigheit av bokstavar i fil


#!/usr/bin/python3
from pathlib import Path
from sys import argv


def get_text_from_file(filename):
    filepath = Path(filename)

    with filepath.open("r") as file:
        text = file.read()
    return text


def print_dict_in_columns(d, width=80, spacer=2):
    terms = []
    output = ""
    
    for k, v in d.items():
        terms.append(f'{k}: {v}%')
    maxlen = len(max(terms, key=len))
    cols = width // (maxlen + spacer)
    rows = round(len(terms) / cols) + 1

    for row in range(rows):
        for col in range(cols):
            n = (rows * col) + row
            if n < len(terms):
                output += terms[n].ljust(maxlen + spacer)
        output += "\n"
    print(output)


def main():
    usage = """Usage: python letter_freq.py file.txt"""
    
    if len(argv) != 2:
        print(usage)
        quit()
        
    file_text = get_text_from_file(argv[1])
    
    letters = [chr(c) for c in range(97, 123)]
    percentage = lambda c: round((file_text.count(c) / len(file_text)) * 100)
    letter_freq = {letter:percentage(letter) for letter in letters}
    
    print_dict_in_columns(letter_freq, width=80)


if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.22
#
#  Krypterer eller dekoder tekst vha. Vigenère-metoden

# Streng med norsk alfabet
ALPHA = "".join([chr(c) for c in range(97, 123)]) + "æøå"


def vigenere_encode_text(text, keyword, decrypt=False):
    keylen = len(keyword)
    processed_text = ""
    unknown_chars = 0
    spaces = 0
    
    text = text.lower()
    keyword = keyword.lower()
    
    for i, char in enumerate(text):
        if char == " ":
            processed_text += " "
            spaces += 1
        
        elif char not in ALPHA:
            # Catchall: Dersom teiknet ikkje er ein liten bokstav eller eit mellomrom
            processed_text += char
            unknown_chars += 1
            
        else:
            # Itererer igjennom bokstavane i "keyword".
            keyword_char_iterator = i % keylen - (unknown_chars % keylen)
            
            # Genererer cipher-streng som er like lang som ALPHA
            shift_by = ALPHA.find(keyword[keyword_char_iterator])
            cipher = ALPHA[shift_by:] + ALPHA[:shift_by]

            if decrypt:
                position_in_cipher = cipher.find(char)
                processed_text += ALPHA[position_in_cipher]
            else:
                position_in_alpha = ALPHA.find(char)
                processed_text += cipher[position_in_alpha]
        
    if unknown_chars != 0:
        print(unknown_chars, "teikn vart ikkje oversatt (kun a-å er implementert).")
        
    return processed_text


### Testkjøring (vil ikkje vere likt som boka sidan eg la til æøå i ALPHA)
text = "denne teksten vart oversatt med kodeordet tiger"
key = "tiger"

enc = vigenere_encode_text(text, key)
print(enc)
dec = vigenere_encode_text(enc, key, decrypt=True)
print(dec)
```

```python
#
#    Besvarelse P7.23
#
#  Krypterer og dekoder tekst vha. Playfair-metoden
#    Det norske alfabetet har 29 bokstavar. Det er
#    dessverre eit primtal. Alternativt kan vi velge
#      28 = 4 * 7   --> må gjere ein bokstav ugyldig
#      30 = 5 * 6   --> må ha eit teikn i tillegg til alfabetet.
#    Eg velger å bruke 30, og å dekode med mellomrom som eit eige
#    teikn.

ALPHA = "".join([chr(c) for c in range(97, 123)]) + "æøå"
CIPHER_WIDTH = 5
CIPHER_HEIGHT = 6


def char_pos_in_array(char, array):
    for y, row in enumerate(array):
        if char in row:
            pos_y = y
            pos_x = row.index(char)
            return pos_y, pos_x


def playfair(text, keyword, print_cipher=False):
    # Passer på at lengden av "text" er eit partal
    text += " " if len(text) % 2 != 0 else ""
    
    # Fjerner evt. duplikat i "keyword"
    keyword = "".join(dict.fromkeys(keyword))
    
    # Genererer cipher-matrise
    cipher = keyword + "".join([char for char in ALPHA if char not in keyword]) + " "
    cipher_array = []
    
    for y in range(CIPHER_HEIGHT):
        cipher_array.append(
            [cipher[n] for n in range(y*CIPHER_WIDTH, y*CIPHER_WIDTH + CIPHER_WIDTH)])
    
    if print_cipher:
        for n in cipher_array:
            print(n)

    # Oversetter strengen "text" vha. cipher-matrisa
    processed_text = ""
    for n in range(0, len(text), 2):
        if n <= len(text):
            letter_1 = text[n]
            letter_2 = text[n + 1]
            
            
            x_1, y_1 = char_pos_in_array(letter_1, cipher_array)
            x_2, y_2 = char_pos_in_array(letter_2, cipher_array)
            
            # Dersom letter_1 of letter_2 ligger i same kolonne eller
            # same rad bytter dei simpelthen plass
            if (x_1 == x_2) or (y_1 == y_2):
                letter_1, letter_2 = letter_2, letter_1
            # Bytter posisjon etter vanleg fairplay-metode
            else:
                letter_1 = cipher_array[x_1][y_2]
                letter_2 = cipher_array[x_2][y_1]
            
            processed_text += letter_1 + letter_2
    return processed_text
            

### Testkjøring
text = "denne teksten har blitt kryptert"
keyword = "playfair"

enc = playfair(text, keyword, print_cipher=True)
print(enc)
dec = playfair(enc, keyword)
print(dec)
```

```python
#
#    Besvarelse P7.32
#
#  Rekner ut spenning over kondensator
#  vha parameter frå fila "param.txt"
#  Skriver resultat til   "rc.txt"

from pathlib import Path
from math import exp


def rc_voltage(t, voltage, resistance, capacitance):
    """V across cap: ---res---switch---cap--- in series, 't' after the switch closes."""
    return voltage * (1 - exp(-t / (resistance * capacitance)))


def write_to_file(filename, text):
    filepath = Path(filename)
    
    with filepath.open("w") as file:
        file.write(text)
        

def get_lines_from_file(filename, wanted_lines):
    filepath = Path(filename)

    with filepath.open("r") as file:
        lines = file.readlines()
    return (float(val) for n, val in enumerate(lines) if n <= wanted_lines - 1)


def main():        
    # Testkjøring, finner 100 verdiar og skriver til fil
    steps = 100
    inputfile = "params.txt"
    outputfile = "rc.txt"
    
    B, R, C, T_0, T_1 = get_lines_from_file(inputfile, 5)
    
    step = (T_1 - T_0) / steps
    lines = []
    for t in range(int(T_0), int(T_1)):
        v = rc_voltage(t, B, R, C)
        lines.append(v)
        if t % 100 == 0:
            print("t:", t, "--->", v)
            
    write_to_file(outputfile, "\n".join([str(val) for val in lines]))
        
        
if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P7.33
#
#  Finner oppladningstida [ms] til kondensator, frå v = 5% til v = 95%
#  av spenningsverdi for full oppladning, B[V], ut ifrå informasjon frå 
#  fila "rc.txt"

from pathlib import Path


def get_lines_from_file(filename):
    filepath = Path(filename)

    with filepath.open("r") as file:
        lines = file.readlines()
    return lines


def main():
    inputfile = "rc.txt"
    inputvals = get_lines_from_file(inputfile)
    rc_data = [float(val) for val in inputvals]
    
    max_val = rc_data[-1]
    
    # Finner verdiane i "inputvals" som ligger nærmast 5% og 95% av "max_val"
    v_1 = min(rc_data, key=lambda x: abs(x - 0.05 * max_val))
    v_2 = min(rc_data, key=lambda x: abs(x - 0.95 * max_val))
    
    t_1 = rc_data.index(v_1)
    t_2 = rc_data.index(v_2)
    
    rise_time = t_2 - t_1
    print("v_1 [V]:", round(v_1, 3))
    print("v_2 [V]:", round(v_2, 3))
    print("rise time [ms]:", rise_time)
    

if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P8.13
#
#  Operasjonar på multisett

def union(set_1, set_2):
    combined_set = dict.fromkeys(set_1, 0)
    for key in set_1:
        if key in set_2:
            combined_set[key] = set_2[key]
        combined_set[key] += set_1[key]
    return combined_set


def intersection(set_1, set_2):
    combined_set = {}
    for key in set_1:
        if key in set_2:
            combined_set[key] = min([set_1[key], set_2[key]])
    return combined_set


def difference(set_1, set_2):
    combined_set = {}
    for key in set_1:
        if key not in set_2:
            combined_set[key] = set_1[key]
    return combined_set


# To eksempel-multisett
ms_1 = {"bananas": 2,
        "apples": 3,
        "wolverines": 13}

ms_2 = {"bananas": 1,
        "apples": 11,
        "mooses": 2}

print(union(ms_1, ms_2))
print(intersection(ms_1, ms_2))
print(difference(ms_1, ms_2))
```

```python
#
#    Besvarelse P8.17
#
#  Laster ned uformatert data med bruttonasjonalinntekt for land, og
#  lar brukaren skrive navn på land for å få informasjon.

import csv
import requests

# Linken i boka er daud, men eg fann fila på archive.org
CSV_URL = "https://web.archive.org/web/20120208044005/https://www.cia.gov/library/publications//the-world-factbook/rankorder/rawdata_2004.txt"


def fetch_csv_data(url):
    with requests.Session() as s:
        download = s.get(url)
        return download

    
def parse_cia_csv_data(data):
        # Fjerner mellomrom, tab og indeksen for kvart element.
        # (Ugly data requires ugly solutions..)
        # Returnerer dictionary {"land": pci}
        
        decoded_content = data.content.decode('utf-8')
        cr = csv.reader(decoded_content.splitlines(), delimiter='$')
        as_list = list(cr)
        country_pci_data = { str(s[0].strip()[s[0].find("\t")+1:]) :
                             int(s[1].strip().replace(",", "")) for s in as_list }
        return country_pci_data


def get_input(prompt="", cast=None, condition=None, error_message=None):
    while True:
        try:
            response = (cast or str)(input(prompt))
            assert condition is None or condition(response)
            return response
        except:
            print(error_message or "Invalid input, try again.")


def main():
    raw_country_pci_data = fetch_csv_data(CSV_URL)
    country_pci_data = parse_cia_csv_data(raw_country_pci_data)

    print("Input country name to see PCI (per capita income). [q]uit exits the program")
    while True:
        user_input = get_input(prompt="Country: ", cast=str)
        
        if user_input in ["q", "quit"]:
            break
        elif user_input in country_pci_data.keys():
            print(f"Per capita income of {user_input}: {country_pci_data[user_input]}")
        else:
            print("Not a country/no data for country.")
    print("Quitting")

            
if __name__ == "__main__":
    main()
```

```python
#
#    Besvarelse P8.18
#
#  Forbetring av programmet i P8.17.
#    - laster ned data og lagrer som pci_data.json
#    - TUI implementert med curses
#    - øyeblikkelig begrensing av alternativer vha. søkeord

import csv
import requests
import json
import curses
from curses import textpad
import os

CSV_URL = "https://web.archive.org/web/20120208044005/https://www.cia.gov/library/publications//the-world-factbook/rankorder/rawdata_2004.txt"


def fetch_csv_data(url):
    with requests.Session() as s:
        download = s.get(url)
        return download

    
def parse_cia_csv_data(data):
        # Fjerner mellomrom, tab og indeksen for kvart element.
        # (Ugly data requires ugly solutions..)
        # Returnerer dictionary {"land": pci}
        
        decoded_content = data.content.decode('utf-8')
        cr = csv.reader(decoded_content.splitlines(), delimiter='$')
        as_list = list(cr)
        country_pci_data = { str(s[0].strip()[s[0].find("\t")+1:]) :
                             int(s[1].strip().replace(",", "")) for s in as_list }
        return country_pci_data

    
def save_dict_to_json(filename, data_dict):
    with open(filename, "w") as file:
        file.write(json.dumps(data_dict))


def inp_field_keyevent(stdscr, inp_field):
    key = stdscr.getch()
    char = chr(key)
    
    if key == (curses.KEY_BACKSPACE or 127):           # Backspace
        if len(inp_field) >= 1:
            inp_field = inp_field[:-1]
    elif key == curses.KEY_RESIZE:
        pass
    elif char.isalpha():                               # Bokstav
        inp_field += char
    elif char in [",", ".", " ", "(", ")"]:            # Lovlege teikn
        inp_field += char

    elif key == curses.KEY_ENTER or key in [10, 13]:   # ENTER
        if inp_field.lower() in ["q", "quit"]:         # avslutt dersom "quit"
            quit()
        else:
            inp_field = ""
            
    return inp_field


def reduce_dict_by_key_search(orig_dict, search):
    matching_keys = [key for key in orig_dict
                     if search.lower() in key.lower()]
    return {key: orig_dict[key] for key in matching_keys}


def print_dict_search_menu(stdscr, inp_field, data):
    h, w = stdscr.getmaxyx()
    stdscr.clear()
    
    # Tekst på toppen
    stdscr.addstr(0, 2, "Per Capita Income search - [q]uit to close")
    
    # Kolonnenavn
    stdscr.addstr(2, 2,    "COUNTRY")
    stdscr.addstr(2, w//2, "Per Capita Income [$], 2004, cia.gov")
    
    # Søkefelt
    stdscr.addstr(h-1, 2, "Input country ->>")
    stdscr.addstr(h-1, 20, inp_field)
    
    # Land og bruttonasjonalinntekt
    curses.textpad.rectangle(stdscr, 1, 1, h-2, w-1)
    avail_h = h-5
    avail_w = w-2
    keys = list(data.keys())
    for i in range(min(avail_h, len(data))):
        # Land
        stdscr.addstr(i+3, 2, keys[i])
        
        # Bruttonasjonalinntekt
        stdscr.addstr(i+3, w//2, str(data[keys[i]]))

    stdscr.refresh()


def main(stdscr):
    global country_pci_data
    pci_data_file = "pci_data.json"
    
    ### Setup
    inp_field = ""
    curses.curs_set(0)
    h, w = stdscr.getmaxyx()
    stdscr.clear()
    min_h, min_w = 4, 30
    
    ### Hent data
    # Dersom pci_data.json allereie eksisterer, spør brukaren om den skal oppdaterast
    if os.path.isfile(pci_data_file):
        print_str = f"{pci_data_file} already exists, update? [y/n]"
        str_offset = len(print_str) // 2
        stdscr.addstr(h//2, w//2 - str_offset, print_str)
        textpad.rectangle(stdscr, h//2 - 1, w//2 - str_offset - 1,
                                  h//2 + 1, w//2 + str_offset + 1)
        user_choice = chr(stdscr.getch())
        
        if user_choice.lower() == "y":
            # Henter ny data
            stdscr.addstr(h//2 + 3, w//2 - 8, "fetching data ...")
            stdscr.refresh()
            raw_country_pci_data = fetch_csv_data(CSV_URL)
            country_pci_data = parse_cia_csv_data(raw_country_pci_data)
        else:
            # Bruk data frå pci_data.json
            with open(pci_data_file, "r") as file:
                data = file.read()
                country_pci_data = json.loads(data)
    else:
        # Ingen data eksisterer, laster ned
        raw_country_pci_data = fetch_csv_data(CSV_URL)
        country_pci_data = parse_cia_csv_data(raw_country_pci_data)
    
    # Skriv til pci_data_file
    save_dict_to_json(pci_data_file, country_pci_data)

        
    ### Hovedloop
    while True:
        # Velg land som skal visast på grunnlag av søkeord
        visible_data = reduce_dict_by_key_search(country_pci_data, inp_field)

        # Printer menyen
        print_dict_search_menu(stdscr, inp_field, visible_data)

        # Venter på, og tolker evt. input frå brukaren
        inp_field = inp_field_keyevent(stdscr, inp_field)
        
         
            
if __name__ == "__main__":
    curses.wrapper(main)
```

```python
#
#    Besvarelse P8.19
#
#

import os

def word_freq_in_file(filename):
    with open(filename, "r") as file:
        data = file.read().split()
    
    word_freq = {}
    for word in set(data):
        word_freq[word] = data.count(word)
    
    return word_freq


def get_input(prompt="", cast=None, condition=None, error_message=None):
    while True:
        try:
            response = (cast or str)(input(prompt))
            assert condition is None or condition(response)
            return response
        except:
            print(error_message or "Invalid input, try again.")

            
def dict_print(dictionary):
    for key, val in dictionary.items():
        print(key, val)
            
            
def main():
    cached_files = {}
    while True:
        filename = get_input(prompt="Enter filename: ", cast=str)
        
        if filename in cached_files.keys():
            dict_print(cached_files[filename])
        elif os.path.exists(filename):
            word_freq = word_freq_in_file(filename)
            dict_print(word_freq)
            cached_files[filename] = word_freq
            print("(File was cached)")
        else:
            print("Filename does not exist")
            
            
if __name__ == "__main__":
    main()
```
